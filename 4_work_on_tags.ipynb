```python
import numpy as np
import pandas as pd
```

```python
df_tag = pd.read_csv('input_data\\tag.csv', delimiter=',')
df_tag.head(16)
```

```python
df_tag.shape
```

```python
# Find columns with NaN values
# Count NaN values for each column
nan_counts = df_tag.isna().sum()

# Filter and print only the columns with NaN values and their counts
nan_columns_counts = nan_counts[nan_counts > 0]
nan_columns_counts
```

```python
df_tag = df_tag.dropna()
```

```python
import nltk
nltk.download('names')
```

```python
from nltk.corpus import names

# Charger les prénoms
name_set = set(names.words())

# Fonction pour vérifier si un tag contient un prénom
def contains_name(tag):
    return any(name in tag.split() for name in name_set)

# Drop les lignes qui contiennent un prénom
df_tag = df_tag[~df_tag['tag'].apply(contains_name)]
df_tag
```

```python
df_tag.dtypes
```

```python
# Convertir la colonne 'timestamp' en datetime si nécessaire
df_tag['timestamp'] = pd.to_datetime(df_tag['timestamp'])

# Calculer l'âge en années
current_date = pd.Timestamp.now()
df_tag['age'] = round((current_date - df_tag['timestamp']).dt.days / 365.25, 0)

# Droper la colonne 'timestamp'
df_tag = df_tag.drop(columns=['timestamp'])
```

```python
df_tag['age'] = round(df_tag['age'],0)
df_tag
```

```python
df_tag.to_csv('output_data/384k_tags_wo_names_4.csv', index=False)
```
